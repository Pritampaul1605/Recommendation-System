# Find top 50 popular books (popularity based rocommendation system)

# 

```python
# marge book_data and rating data set

data = book_data.merge(rating, on = 'ISBN')
data.head()
```

```python
data.shape
```

```python
# create a data frame based on the total rating of each book

popular_book = data.groupby('Book-Title').count()[['Book-Rating']]
popular_book
```

```python
popular_book.rename({'Book-Rating':'total_rating'},axis = 1,inplace = True)
```

```python
# cerate a new column based on the average rating of each book

popular_book['avg_rating'] = data.groupby('Book-Title').mean()[['Book-Rating']]
popular_book
```

```python
# sort the books based on the average rating

popular_book.sort_values('avg_rating', ascending = False, inplace = True)
```

```python
#take only those books which get more than 250 rating and find the top 50 books which have more average rating

popular_book = popular_book[popular_book['total_rating'] >=250].head(50)
popular_book
```

```python
# marge popular data with data and take all the infomative columns of 50 books

popular_books = popular_book.merge(data, on = 'Book-Title')[['Book-Title','Book-Author','Image-URL-M','total_rating','avg_rating']]
popular_books
```

# collaborative filtering

```python
data 
```

```python
x = data.groupby('User-ID').count()['Book-Rating'] > 200
educated_user = x[x].index
```

```python
filtered_data = data[data['User-ID'].isin(educated_user)]
filtered_data
```

```python
y = filtered_data.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book = y[y].index
```

```python
final_data = filtered_data[filtered_data['Book-Title'].isin(famous_book)]
final_data
```

```python
pt = final_data.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating').fillna(0)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score
```

```python
similarity_score
```

```python
def recommender(book_name) :
    id = np.where(pt.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[id])), key= lambda x:x[1], reverse = True)[1:6]
    
    for i in similar_item :
        print(pt.index[i[0]])
```

```python
recommender('1984')
```
